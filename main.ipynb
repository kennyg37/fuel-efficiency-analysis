```python
# YOUR CODE HERE
coData = df['co2']

sns.set(style="whitegrid")

sns.histplot(coData, bins=20, color='red', edgecolor='black')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide Emissions')
plt.show()
```

```python
# YOUR CODE HERE

cylinders = df['cylinders']

sns.histplot(data=cylinders, bins=20, color='violet', edgecolor='black')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.title('Cylinders Frequency')


plt.show()
print(cylinders)
```

```python
# YOUR CODE HERE

Combfuel= df['comb']
sns.histplot(data=df, x='comb', color='yellow', edgecolor='black', bins=20)
plt.xlabel('Combined Fuel Efficiency')
plt.ylabel('Frequency')
plt.title('Distribution of Combined Fuel Efficiency')
plt.show()
```
